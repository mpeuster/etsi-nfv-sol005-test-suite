# ETSI SOL005 vs. OSM rel. 4 evaluation

Execution: `pytest --durations=0`
Results in: `durations.result`

```python
# parse input file an create a nice pandas frame
import pandas as pd
import seaborn as sns

data = list()
with open("durations.result", "r") as f:
    for l in f.readlines():
        if "call" not in l:
            continue
        parts = list(map(str.strip, filter(len, l.split(" "))))
        d = {
            "name": parts[2].split("::")[-1],
            "time": float(parts[0].replace("s", ""))
        }
        data.append(d)
df = pd.DataFrame(data)
print(df.info())
print(df.describe())
display(df.head())
```

```python
# analyse tests
g = df.groupby(by="name").describe()
display(g.sort_values("name"))
```
